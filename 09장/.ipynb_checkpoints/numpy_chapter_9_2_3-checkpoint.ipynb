표본 분산(sample variance)

- 데이터와 표본 평균간의 거리의 제곱의 평균이다. 표본 분산이 작으면 데이터가 모여있는 것이고 크면 흩어져 있는 것이다. 수학 기호로는 $s^2$이라고 표시하며 다음과 같이 계산한다.

$$ s^2 = \dfrac{1}{N}\sum_{i=1}^N (x_i - \bar{x})^2 $$

표본 표준편차(sample standard variance)

- 표본 분산의 양의 제곱근 값이다. $s$이라고 표시한다.

$$ s = \sqrt{s^2} $$

### 표본 분산 (var) 함수와 표준 편차(std) 함수

```python
a = np.array([[3,7,5],[8,4,3],[2,4,9]])
```

```python
a
```

```python
np.var(a)
```

```python
m = np.mean(a)
```

```python
m
```

```python
np.sum(np.power((a - m), 2))/ len(a.flatten())
```

```python
np.sqrt(np.var(a))
```

```python
np.std(a)
```

## 예제 9-2-3-2  표본 분산 및 표준편차 알아보기 

```python
Image("C:\\Users\\MG\\Desktop\\칼무리\\넘파이9장/K-009.png")
```

```python
f = np.array([1,4,3,8,9,2,3])
```

```python
np.var(f,ddof=1) # ddof 는 n-1을 뜻함.
```

```python
m = np.mean(f)
```

```python
m
```

```python
np.sum(np.power((f - m), 2))/ (len(f.flatten()) - 1)
```

```python
np.std(f,ddof=1)
```

```python
np.sqrt(np.var(f, ddof=1))
```

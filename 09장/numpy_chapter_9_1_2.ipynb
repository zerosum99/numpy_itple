## 예제 9-1-2-1  정규분포 함수 알아보기 

normal 함수 

ex normal(mu, sigma, n)

- 평균(mu)과 표준편차(sigma)를 이용하여 1000개의 원소를 갖는 정규분포 배열을 만들어줍니다.

```python
mu, sigma = 0, 1
```

```python
s = np.random.normal(mu, sigma, 1000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(s)

plt.show()
```

```python
t = np.random.normal(1,2, 1000)
```

```python
plt.hist(t)

plt.show()
```

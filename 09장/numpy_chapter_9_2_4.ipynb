### 상관계수

확률변수의 절대적 크기에 영향을 받지 않도록 공분산을 단위화시킨 것이다. 즉, 공분산에 각 확률변수의 분산을 나눠주었다.

```python
a = np.array([5,6,7,8])
```

```python
a
```

```python
b = np.array([1,3,4, 5])
```

```python
b
```

```python
np.corrcoef(a,b)
```

```python
np.cov(a,b)
```

```python
ma = np.mean(a)
```

```python
mb = np.mean(b)
```

```python
np.sum((a -ma) * (b-mb)) / (len(a) - 1)
```

```python
np.std(a,ddof=1) * np.std(b, ddof=1)
```

```python
(np.sum((a -ma) * (b-mb)) / (len(a) - 1) ) / (np.std(a,ddof=1) * np.std(b, ddof=1))
```

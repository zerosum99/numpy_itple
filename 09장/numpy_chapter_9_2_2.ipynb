중앙값과 평균

- 중앙값 : 주어진 원소들의 중간에 있는값
- 평균 : 주우진 원소들을 더한 후 원소들의 갯수로 나누어진 결과값
$$ \bar{x} = \dfrac{1}{N}\sum_{i=1}^N x_i $$

```python
d = np.array([3,4,2,3,8,9,5])
```

```python
np.median(d)
```

```python
np.sort(d)
```

```python
e = np.array([3,4,2,3,8,9,5,7])
```

```python
np.median(e)
```

```python
np.sort(e)
```

```python
d.mean()
```

```python
np.mean(d)
```

```python
np.average(d)
```

```python
f = np.reshape(e, (4,2))
```

```python
np.mean(f, axis=0)
```

```python
np.mean(f, axis=1)
```

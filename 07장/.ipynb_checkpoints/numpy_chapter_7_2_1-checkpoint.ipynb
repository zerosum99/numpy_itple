## 예제 7-2-1-1 선형함수 알아보기

```python
def spines() :
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
```

```python
def identity(x) :
    return x
```

```python
x = np.arange(-10,11)
```

```python
y = identity(x)
```

```python
x == y
```

```python
fig = plt.figure(figsize=(8,6))

plt.plot(x,y)
spines()
plt.show()
```

```python
z = np.arange(-10,11) + 3 * x
```

```python
fig = plt.figure(figsize=(8,6))

plt.plot(z,y)
spines()
plt.show()
```

## 예제 7-2-1-2 비선형함수 알아보기

```python
def step(x) :
    return np.where(x>0, 1,0)
```

```python
x = np.arange(-5,5,0.1)
```

```python
y = step(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
def sigmoid(x) :
    return 1 /(1 + np.exp(-x))
```

```python
x = np.arange(-5,5, 0.1)
```

```python
y = sigmoid(x)
```

```python
x.shape, y.shape
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
x = np.arange(-5,5,0.1)
```

```python
y = np.tanh(x)
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
def relu(x) :
    return np.maximum(0,x)
```

```python
x = np.arange(-5,5,0.1)
```

```python
y = relu(x)
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
def softmax(x) :
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.arange(1,5)
```

```python
y = softmax(x)
```

```python
y
```

```python
np.sum(y)
```

```python
ratio = y
labels = y
 
plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```

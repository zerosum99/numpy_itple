## 예제 7-1-2-1  그래프에 설명 달기

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(1,8,0.1)
```

```python
y = np.sin(x)
```

```python
fig = plt.figure(figsize=(8,6))
plt.title(' SIN Curve ')
plt.xlabel(" x axis ")
plt.ylabel(" x axis ")

plt.plot(x,y, color='0.75', linestyle='solid')
plt.plot(x+0.3, y,  color='#00ffff', linestyle='dashed')
plt.plot(x+0.6, y,  color=(0.35, 0.56, 0.65), linestyle='dashdot')
plt.plot(x+0.9, y,  color=(0.35, 0.56, 0.65, 0.50), linestyle='dashed')
plt.show()
```

## 예제 7-1-2-2 그래프에 해당하는 범례 추가하기 

```python
fig = plt.figure(figsize=(8,6))
plt.title(' SIN Curve ')
plt.xlabel(" x axis ")
plt.ylabel(" x axis ")

plt.plot(x,y, color='0.75', linestyle='solid', label='Solid')
plt.plot(x+0.3, y,  color='#00ffff', linestyle='dashed', label='Dashed 1')
plt.plot(x+0.6, y,  color=(0.35, 0.56, 0.65), linestyle='dashdot', label='DashHot')
plt.plot(x+0.9, y,  color=(0.35, 0.56, 0.65, 0.50), linestyle='dashed', label='Dashed 2')

plt.legend()
plt.show()
```

## 예제 7-1-3-1  그래프 내의 좌표 꾸미기 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [1,2,3,4]
```

```python
y = [20, 21, 20.5, 20.8]
```

```python
plt.plot(x,y) 
plt.show()
```

```python
plt.plot(x,y, linestyle='dashed', marker='o', color='green') 

plt.xlim(0.5, 4.5)
plt.ylim(19.5, 21.5)
plt.show()
```

```python
plt.plot(x,y, linestyle='dashed', marker='o', color='green') 

plt.xlim(0.5, 4.5)
plt.xticks(x)
plt.ylim(19.5, 21.5)
plt.yticks(y)

plt.show()
```

```python
x = [0,1,2]
```

```python
y = [90, 40, 45]
```

```python
labels = ['High', 'Low', 1234567]
```

```python
plt.plot(x,y,'r')
plt.xticks(x, labels, rotation='vertical')

plt.show()
```

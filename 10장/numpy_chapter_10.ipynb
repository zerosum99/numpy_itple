```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0, 10, 0.01)
y = 3*t + 5
y_noise = y + np.random.randn(len(y))
```

<b>그래프 그릴 모듈을 import 하고 입력값을 1차원 배열로 만든다

이를 poly1d함수에 넣어서 다항식을 만들어서 적정한 방정식을 찾는다

```python
fp1 = np.polyfit(t, y_noise, 1)
f1 = np.poly1d(fp1)
```

```python
fp1
```

```python
f1
```

<b>두 개의 그래프를 만들기 위해 subplots 함수로 행 1을 지정하고 열에 2를 지정한 후에 실행을 하면

```python
fig, axes = plt.subplots(1,2, figsize=(10, 6))
axes[0].plot(t, y_noise, label='noise', color='y')
axes[0].plot(t, y, ls='dashed', lw=3, color='b', label='original')
axes[0].legend()
axes[0].grid()
axes[1].plot(t, y_noise, label='noise', color='y')
axes[1].plot(t, f1(t), lw=2, color='r', label='polyfit')
axes[1].legend()
axes[1].grid()
plt.show()
```

## 예제 10-1-2-2   lstsq 함수로 연립 방정식 풀기

###  lstsq 최소자승 해 풀기

lstsq 명령은 행렬  𝐴
 와  𝑏  모두 인수로 받고 뒤에서 설명할 최소자승문제(least square problem)의 답 x, 잔차제곱합(residual sum of squares) resid, 랭크(rank) rank, 특잇값(singular value) s를 반환한다. 미지수와 방정식의 개수가 같고 행렬  𝐴의 역행렬이 존재하면 최소자승문제의 답과 선형 연립방정식의 답이 같으므로 lstsq 명령으로 선형 연립방정식을 풀 수도 있다. 최소자승문제, 랭크, 특잇값에 대해서는 뒤에서 자세히 설명할 것이다.


lstsq 명령을 사용하는 것이 inv 명령을 사용하는 것보다 수치오차가 적고 코드도 간단하므로 선형 연립방정식의 해를 구할 때도 lstsq 명령을 사용하는 것을 권장한다.

## 예제 10-2-2-2  중앙 차분으로 미분 알아보기

```python
x= 10
```

```python
def f(x):
    return x**2
```

<b>하나의 변수와 변수를 제곱하는 함수를 정의

<b>수치 미분을 처리하는 함수를 정의하고, 매개 변수는 함수와 입력 데이터를 받는다 (변화량은 작은 값으로 지정)

중앙차분은 첫 번째 함수에 인자로는 입력데이터와 변화량을 더하고 두 번째 함수에는 입력데이터에 변화를 뺀 후에 함수의 실행결과를 뺄셈으로 처리한다

분모인 변화량은 2를 곱해서 전체의 변화율을 계산한다

```python
def numerical_differentiation(f, x):    # 수치미분
    delta_x = 1e-7    # 0.0000001
    return (f(x+delta_x) - f(x-delta_x)) / (delta_x*2)  # 중앙 차분   
```

<b>중앙차분으로 미분을 하는 함수의 인자로 함수 f와 입력값 x를 전달해서 미분 값을 계산한다

```python
nd = numerical_differentiation(f, x)
```

```python
nd
```

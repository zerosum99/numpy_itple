#### * 함수를 위한 그래프 그리기

```python
# 그래프의 중심을 특정 위치로 이동하려면 좌표출에 해당하는 spines를 이동해야 합니다. 
# 이동할 때마다 spines 처리하지 않기 위해 하나의 함수 spines를 정의해서 사용합니다.
# spines는 4개의 좌표축을 가지므로 오른쪽과 위를 표시하지 않도록 색깔을 주지 않습니다. 
# 아래와 좌측에만 좌표 측을 중심으로 이동하고 레이블을 지정합니다.

def spines() :
    ax = plt.gca() # 현재의 축 구하기
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data', 0))
    
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
```

```python
x = np.arange(-10, 13)
y = 3*x + 3
plt.plot(x, y)
spines()
plt.show()
```

```python
x = np.arange(-10, 13)
y = -3*x + 3
plt.plot(x, y)
spines()
plt.show()
```

```python
# 항등함수

y = x
plt.plot(x, y)
spines()
plt.show()
```

### 3. 활성화 함수 그래프로 그려보기

>#### Quiz  2
>활성화 함수를 선형 함수로 사용하지 않는 이유를 적어주세요.


```python
## 계단함수

def step(x) :
    return np.where(x>0, 1, 0)

x = np.arange(-5, 5, 0.1)
y = step(x)
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 시그모이드 함수 : 자연상수를 지수로 사용해 작은 값과 큰 값을 특정 범위 내의 값으로 처리. 0과 1사이로 변환

def sigmoid(x) :
    return 1/(1 + np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 하이퍼볼릭 탄젠트 함수 : 자연상수를 분모에는 더하고 분자에는 빼는 구조. -1과 1사이의 값을 0을 중심으로 대칭형태로 변환

def htan(x) :
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = htan(x)
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 렐루 함수 : 입력값이 0보다 크면 그 값을 그대로 반환하고, 0보다 작거나 같으면 0으로 출력. 
# 기울기의 미분을 역으로 전파 될 때 기울기의 값이 사라지는 것을 방지하기 위해 많이 사용됨.

def relu(x) :
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)
y = relu(x)
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

>#### Quiz  3
>역전파(Backpropagation)란 무엇이고 왜 딥러닝에서 왜 중요한가?

```python
# 소프트맥스 함수 : 0보다 크고 1보다 작은 n개 노드 출력. n개별 확률값을 출력하여 분류를 구하는 출력층에 주로 사용.

def softmax(x) :
    return np.exp(x) / np.sum(np.exp(x))

x = np.arange(1, 5)
y = softmax(x)

np.sum(y) # 1입니다.

ratio = y
labels = y

plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```
